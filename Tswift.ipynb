```python
unique_words_df = pd.DataFrame(columns=['Word', 'Count'])
# Iterate over each row in the 'text_column' of the DataFrame
for row in df['lyric']:
    words = row.split()  # Split the string into a list of words
    unique_words = set(words)  # Get the unique words in the list

    # Iterate over each unique word
    for word in unique_words:
        if word in unique_words_df['Word'].values:
            # If the word already exists in the DataFrame, increment its count
            unique_words_df.loc[unique_words_df['Word'] == word, 'Count'] += 1
        else:
            # If the word doesn't exist, add it to the DataFrame with count 1
            unique_words_df = unique_words_df.append({'Word': word, 'Count': 1}, ignore_index=True)

# Print the resulting DataFrame with unique words and their counts
print(unique_words_df)

    
```

```python
sorted_df = unique_words_df.sort_values(by='Count', ascending=False)

# Print the sorted DataFrame
print(sorted_df)
```

```python
sorted_df = unique_words_df.sort_values(by='Count', ascending=False)

# Save the sorted DataFrame to a CSV file named 'unique_words.csv'
sorted_df.to_csv('unique_words.csv', index=False)

# Print a confirmation message
print("DataFrame saved to unique_words.csv")
```

```python
spotify=pd.read_csv("tswift/Taylor_Swift_Spotify_Data.csv")
spotify = spotify.drop(['artist_name','artist_id','album_id','album_type','album_release_date_precision','track_id','analysis_url','track_href','is_local','track_preview_url','track_uri','external_urls.spotify','album_release_date','album_release_year'],axis=1)
spotify=spotify.copy()
```

```python
spotify1=spotify.drop(['disc_number','explicit','track_name','type','album_name','key_mode','mode_name'],axis=1)
```

```python
encoded_data = pd.get_dummies(spotify['key_name'], prefix='key')
data_encoded = pd.concat([spotify1, encoded_data], axis=1)
for col in data_encoded.columns:
    print(col)
```

```python
X=spotify1.drop(['track_number','key_name'],axis=1)
X=X.values
y=spotify['track_number']
```

```python
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.3, random_state=0)
```

```python
clf = KNeighborsClassifier(n_neighbors=15)
```

```python
clf.fit(X_train, y_train)
```

```python
print("Test set predictions:",clf.predict(X_test))
```

```python
print("Test set accuracy:{:2f}".format(clf.score(X_test,y_test)))
```
